```python
import csv
from matplotlib import pyplot as plt
```

```python
def file_open(sample_thickness):
    with open(str(sample_thickness) + ' per.is_metal_RawData/Specimen_RawData_1.csv') as f:
        data = list(csv.reader(f, delimiter=','))
        stress = []
        strain = []
        for row in data[2:]:
            stress.append(float(row[3]))
            strain.append(float(row[4]))
    return stress, strain
```

```python
stress_20, strain_20 = file_open(20)
stress_40, strain_40 = file_open(40)
stress_60, strain_60 = file_open(60)
fig = plt.figure(figsize=(10, 10))
axis = fig.add_subplot(111)
axis.set(title='Stress vs Strain', xlabel='Strain', ylabel='Stress (MPa)')
axis.plot(strain_20, stress_20, label='20 mm sample')
axis.plot(strain_40, stress_40, label='40 mm sample')
axis.plot(strain_60, stress_60, label='60 mm sample')
plt.legend()
plt.show()
```
